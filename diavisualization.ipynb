```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
diabetes = pd.read_csv(r'C:\Users\molab\Documents\Learning python\diabetes\cleaned_diabetes.csv')
diabetes['Gender'] = diabetes['Gender'].apply(lambda x: 0 if x == 'F' else 1)
print(diabetes.head())
```

```python
sns.countplot(x='CLASS', data=diabetes)
plt.title("Distribution of CLASS")
plt.show()
```

```python
sns.boxplot(x='CLASS', y='HbA1c', data=diabetes)
plt.title("HbA1c vs CLASS")
plt.show()
sns.boxplot(x='CLASS', y='Cr', data=diabetes)
plt.title("Cr vs CLASS")
plt.show()
sns.boxplot(x='CLASS', y='BMI', data=diabetes)
plt.title("BMI vs CLASS")
plt.show()
```

```python
corr = diabetes.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation between Features")
plt.show()
```

```python
sns.countplot(x='Gender', hue='CLASS', data=diabetes)
plt.title("Gender vs CLASS")
plt.show()
```

```python
diabetes['AgeGroup'] = pd.cut(diabetes['AGE'], bins=[0, 20, 40, 60, 80], labels=['0-20', '21-40', '41-60', '61-80'])

sns.countplot(x='AgeGroup', hue='CLASS', data=diabetes)
plt.title("Age Group vs CLASS")
plt.show()
```
